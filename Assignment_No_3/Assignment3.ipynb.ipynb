## Bar plot

```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


try:
    df = pd.read_csv('Admission_Predict.csv')
except FileNotFoundError:
    print("Error: 'Admission_Predict.csv' not found. Please make sure the file exists in the current directory or provide the correct path.")
except pd.errors.ParserError:
    print("Error: Could not parse the CSV file. Please check its format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")



try:
    plt.figure(figsize=(10, 6))
    sns.barplot(x='GRE Score', y='Chance of Admit ', data=df, ci=None)
    plt.title('GRE Score vs. Chance of Admit')
    plt.xlabel('GRE Score')
    plt.ylabel('Chance of Admit')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
except NameError:
    print("Error: 'df' is not defined. Make sure the DataFrame is loaded successfully before plotting.")
except KeyError:
    print("Error: 'GRE Score' or 'Chance of Admit ' column not found in the DataFrame.")
except Exception as e:
    print(f"An unexpected error occurred during plotting: {e}")
```

## Line Chart

```python

import matplotlib.pyplot as plt
try:
    plt.figure(figsize=(10, 6))
    sns.lineplot(x='CGPA', y='Chance of Admit ', data=df)
    plt.title('CGPA vs. Chance of Admit')
    plt.xlabel('CGPA')
    plt.ylabel('Chance of Admit')
    plt.show()
except NameError:
    print("Error: 'df' is not defined. Make sure the DataFrame is loaded successfully before plotting.")
except KeyError:
    print("Error: 'CGPA' or 'Chance of Admit ' column not found in the DataFrame.")
except Exception as e:
    print(f"An unexpected error occurred during plotting: {e}")
```

## Pie chart

```python

import matplotlib.pyplot as plt

try:
    research_counts = df['Research'].value_counts()
    labels = ['No Research (0)', 'Research (1)']
    plt.figure(figsize=(8, 6))
    plt.pie(research_counts, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.title('Distribution of Students with and without Research Experience')
    plt.axis('equal')
    plt.show()
except NameError:
    print("Error: 'df' is not defined. Make sure the DataFrame is loaded successfully before plotting.")
except KeyError:
    print("Error: 'Research' column not found in the DataFrame.")
except Exception as e:
    print(f"An unexpected error occurred during plotting: {e}")
```

## Box Chart

```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

try:
    df = pd.read_csv('Admission_Predict.csv')
except FileNotFoundError:
    print("Error: 'Admission_Predict.csv' not found. Please make sure the file exists in the current directory or provide the correct path.")
except pd.errors.ParserError:
    print("Error: Could not parse the CSV file. Please check its format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")


try:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Chance of Admit ', y='GRE Score', data=df)
    plt.title('GRE Score vs. Chance of Admit')
    plt.xlabel('Chance of Admit')
    plt.ylabel('GRE Score')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
except NameError:
    print("Error: 'df' is not defined. Make sure the DataFrame is loaded successfully before plotting.")
except KeyError:
    print("Error: 'GRE Score' or 'Chance of Admit ' column not found in the DataFrame.")
except Exception as e:
    print(f"An unexpected error occurred during plotting: {e}")
```
