```python
# Count the number of zeros in each column
zero_counts = (dataset == 0).sum()
print("Number of zeros in each column:\n", zero_counts)
```

**f)indexing and seleting and sorting data**

```python

# Sort the selected data by 'Year' in ascending order
sorted_data = select_data.sort_values(by='Year', ascending=True)

# Display the sorted data
sorted_data
```

**h)count unique values of data , format of each column, converting variable data type (e.g.from long to short, vice versa)**

```python
# Count unique values in each column
unique_counts = dataset.nunique()
print("Number of unique values in each column:\n", unique_counts)
```

```python
# Check if 'Year' column exists
if 'Year' in dataset.columns:
    # Convert 'Year' column to int64 (long int)
    dataset['Year'] = dataset['Year'].astype('int64')
    print("\nData types after conversion:")
    print(dataset.dtypes)
else:
    print("'Year' column not found in the dataset.")
```
